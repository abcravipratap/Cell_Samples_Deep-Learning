```python
!pip install opendatasets
import opendatasets as opd
opd.download("https://www.kaggle.com/datasets/ammaraahmad/top-10-machine-learning-datasets")
import pandas as pd
data= pd.read_csv("/content/top-10-machine-learning-datasets/cell_samples.csv")
data.head()
```

```python
data.info()
```

```python
data.drop(data.index[data['BareNuc']=='?'],inplace=True)
```

```python
data.info()
```

```python
data['BareNuc']=data['BareNuc'].astype('int')
```

```python
data
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
X=data.drop(['Class'],axis=1)
y=data['Class']
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_scaled=sc.fit_transform(X)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y_scaled=le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y_scaled,test_size=0.2,random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(30,activation='relu',input_dim=9))
model.add(Dense(15,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100)
```

```python
y_pred=model.predict(x_test)
y_pred_binary = (y_pred > 0.5).astype(int)
y_pred_binary
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_binary)

```

```python

user_data= [[1,2,3,4,5,6,7,8,9]]
user_data_scaled=sc.transform(user_data)
prediction=model.predict(user_data_scaled)
prediction
final_prediction=(prediction>0.5).astype(int)
final_prediction
```
